```python
## Let plot the graph again
df.plot(subplots=True,figsize=(14,12) )
```

```python
## Line graph is not suitable for non time series dataseet. however we may get some good information from the line graph
```

```python
# let us plot only for some specific column say age
df.Age.plot(figsize=(12,5))
plt.show()
```

```python
## Let us plot with Customerization
```

```python
df.Age.plot(figsize=(12,5), fontsize=20,c="red")
plt.title('Age',loc="right",fontsize=16)
plt.xlabel('No. of passenger')
plt.ylabel('Age')
plt.show()
```

# Let us plot in Histogram

```python
## Histogram plot will show the frequency distribution of the data set
```

```python
df['Age'].value_counts()
```

```python
df['Age'].value_counts(bins=20)
```

```python
df['Age'].plot(kind='hist',figsize=(12,5),fontsize=15,bins=20)
plt.show()
```

```python
# the above plot shows that the majority of the passenger ware in the age range of 17-40
```

```python
df['Age'].plot(kind='hist',figsize=(12,5),fontsize=15,bins=100)
plt.title('Age Distribution', fontsize=17)
plt.xlabel( 'No of passenger',fontsize=17)
plt.ylabel( 'Age',fontsize=17)
plt.show()
```

```python
## Alternative way of ploting histogram
```

```python
plt.figure(figsize=(12,3))
plt.hist(df['Age'],bins=80)
plt.ylabel('Age')

plt.show()
```

```python
## The only difference in plot method and Hist method is that plot method can handle missing values but in hist method 
#may generate error message
```
